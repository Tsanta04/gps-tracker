```python
import time
import json
import random
import ssl
import paho.mqtt.client as mqtt
from datetime import datetime
```

```python
BROKER = "e792f3c61e8648d2bd1822c6d1910301.s1.eu.hivemq.cloud"
PORT = 8883
USERNAME = "GPS_ESP"
PASSWORD = "TsantaMirindra1"

DEVICE_ID = "device001"
CLIENT_ID = "location-subscriber"
TOPIC = "iot/"+DEVICE_ID+"/location"

# Position de départ (ex: Antananarivo / Larissa / etc.)
BASE_LAT = -18.8792
BASE_LNG = 47.5079
```

```python
def on_connect(client, userdata, flags, rc):
    if rc == 0:
        print("Connected to HiveMQ Cloud")
        print(TOPIC)
    else:
        print(f"Connection failed (code {rc})")
```

```python
client = mqtt.Client(client_id=DEVICE_ID)
client.username_pw_set(USERNAME, PASSWORD)
client.tls_set(cert_reqs=ssl.CERT_REQUIRED)

client.on_connect = on_connect
client.connect(BROKER, PORT)
client.loop_start()
```

```python
# ================= TRAJET =================
GAIOPOLIS = (-18.880500, 47.508900)
CENTRAL_SQUARE = (-18.879200, 47.507000)

STEPS = 50          # nombre de points (plus = plus lent et réaliste)
DELAY = 2           # secondes entre chaque position

try:
    for i in range(STEPS + 1):
        # interpolation linéaire
        lat = GAIOPOLIS[0] + (CENTRAL_SQUARE[0] - GAIOPOLIS[0]) * i / STEPS
        lng = GAIOPOLIS[1] + (CENTRAL_SQUARE[1] - GAIOPOLIS[1]) * i / STEPS

        payload = {
            "device_id": DEVICE_ID,
            "latitude": round(lat, 6),
            "longitude": round(lng, 6),
            "timestamp": datetime.utcnow().isoformat()
        }

        client.publish(TOPIC, json.dumps(payload), qos=1)
        print("Sent:", payload)

        time.sleep(DELAY)

except KeyboardInterrupt:
    print("\nStopped by user")

finally:
    client.loop_stop()
    client.disconnect()
```
