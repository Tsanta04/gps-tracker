```python
import time
import json
import random
import ssl
import paho.mqtt.client as mqtt
from datetime import datetime
```

```python
BROKER = "e792f3c61e8648d2bd1822c6d1910301.s1.eu.hivemq.cloud"
PORT = 8883
USERNAME = "GPS_ESP"
PASSWORD = "TsantaMirindra1"
CLIENT_ID = "location-subscriber"
DEVICE_ID = "device001"
TOPIC = "iot/"+DEVICE_ID+"location"

# Position de d√©part (ex: Antananarivo / Larissa / etc.)
BASE_LAT = -18.8792
BASE_LNG = 47.5079
```

```python
def on_connect(client, userdata, flags, rc):
    if rc == 0:
        client.subscribe(TOPIC)
        print("Connected to HiveMQ Cloud")
    else:
        print(f"Connection failed (code {rc})")
```

```python
def on_message(client, userdata, msg):
    try:
        payload = msg.payload.decode()
        data = json.loads(payload)

        print("\nNew location received")
        print("Topic:", msg.topic)
        print("Device ID:", data.get("device_id"))
        print("Latitude:", data.get("latitude"))
        print("Longitude:", data.get("longitude"))
        print("Timestamp:", data.get("timestamp"))

    except Exception as e:
        print("Error parsing message:", e)
        print("Raw payload:", msg.payload.decode())
```

```python
client = mqtt.Client(client_id=CLIENT_ID)
client.username_pw_set(USERNAME, PASSWORD)
client.tls_set(cert_reqs=ssl.CERT_REQUIRED)

client.on_connect = on_connect
client.on_message = on_message

client.connect(BROKER, PORT)
client.loop_forever()
```
